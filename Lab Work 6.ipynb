```python
def check_even(n):
    if not isinstance(n,int):
        raise TypeError("Input must be an integer")
    if n % 2 != 0:
        raise ValueError("The number is odd")
        return True
try:
    check_even(3)
except (TypeError, ValueError) as e:
    print(e)
```

```python
try:
    age = int(input("Enter your age:"))
    assert age > 18, "Age must be above 18"
    print("Access granted")
except AssertionError as msg:
    print(msg)
```

```python
def check_palindrome(s):
    assert s != "", "Input string cannot be empty"
    
    if s == s[::-1]:
        print("Palindrome")
    else:
        print("Not a palindrome")

check_palindrome("madam")
```

```python
class InsufficientBalanceError(Exception):
    pass

def withdraw(balance, amount):
    if amount > balance:
        raise InsufficientBalanceError("Insufficient balance")
    return balance - amount

try:
    print(withdraw(1000, 500))  
    print(withdraw(1000, 1500))
except InsufficientBalanceError as e:
    print(e)
```

```python
class InvalidEmailError(Exception):
    pass

def validate_email(email):
    if "@" not in email or not (email.endswith(".com") or email.endswith(".org")):
        raise InvalidEmailError("Invalid email format")
    print("Valid email")

try:
    validate_email("test@example.com")
    validate_email("invalid-email")
except InvalidEmailError as e:
    print(e)
```

```python
class HighTemperatureError(Exception):
    pass

def check_temperature(temp):
    if not isinstance(temp, (int, float)):
        raise TypeError("Temperature must be a number")
    
    assert -273 <= temp <= 10000, "Temperature out of realistic range"
    
    if temp > 1000:
        raise HighTemperatureError("Temperature too high")
    
    print("Temperature is acceptable:", temp)

try:
    check_temperature(25)
    check_temperature(1500)
except (TypeError, AssertionError, HighTemperatureError) as e:
    print(e)
```
