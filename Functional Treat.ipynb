```python
### Functional Treat Project

# find factorial
def findFact(num):
    if num < 0:
        return "Invalid number .. please enter positive."
    result = 1

    for i in range(1,num+1):
        result *= i
    return result


def data_summery(data):
    print("Total Data : ",len(data))
    print("Minimum Value:",min(data))
    print("Maximum Value : ",max(data))
    print("Sum of all Value:",sum(data))
    print("Average of value:",sum(data)/len(data));

def filter_above(Data, threshold):
    filtered_obj = filter(lambda x: x > threshold, Data)
    result = list(map(lambda x: x, filtered_obj))
    return result

def dataset(data):
    print("Minimun value:",min(data))
    print("maximum value:",max(data))
    print("sum of all value:",sum(data))
    print("Average of value:",sum(data)/len(data))
        

  

print("Welcomt to the Data Anlyzer and Data Transformer")

while True:
    print("\n Main Menu:")
    print("1.Input Data")
    print("2.Display Data sumay")
    print("3.Calculate Factorial")
    print("4.Filter Data by Threshold")
    print("5.Sort Data")
    print("6.Display Dataset statistics")
    print("7.Exit Program")

    choice = input("Enter your choice:");

    if choice == "1":
        input_data = input("Enter data element with space:");
        data = [int(x) for x in input_data.split(" ")]
        print("Data added successfully.")
        
    elif choice == "2":
        data_summery(data)

    elif choice == "3":
        num = int(input("Enter any number:"))
        print("Factorial is :",findFact(num))

    elif choice == "4":
        thresh = int(input("Enter threshold: "))
        f = filter_above(data, thresh)
        print(f"Values above {thresh}: {f}")

    elif choice == "5":
        print("Choose Sorting Option:")
        print("1.Asceding")
        print("2.Desceding")
        
        option = input("Enter your sorting option:")
        
        if option == "1":
            data.sort()
            print(data)
        
        elif option == "2":
            data.sort(reverse=True)
            print(data)
        else:
            print("Invalid option.")

    elif choice == "6":
        dataset(data)
        
    elif choice == "7":
        print("Thank you for using our system.")
        break;
```
