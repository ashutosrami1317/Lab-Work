```python
text = input("Enter a string: ")
rev = text[::-1]
print("Reversed:", rev)

if text == rev:
    print("It is a palindrome")
else:
    print("It is not a palindrome")
```

```python
text = "Hello World"
print(f"Starts with 'Hello': {text.startswith('Hello')}")
print(f"Ends with 'World': {text.endswith('World')}")

messy_str = "Data123#Science"
clean_str = "".join(char for char in messy_str if char.isalpha())
print(f"Cleaned string: {clean_str}")
print(f"python reversed: {'Python'[::-1]}")
```
