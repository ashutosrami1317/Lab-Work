```python
students = []

while True:
    print("\n===== Student Data Manager =====")
    print("1. Add Student")
    print("2. View Students")
    print("3. Search Student")
    print("4. Delete Student")
    print("5. Show All Subjects")
    print("6. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        print("\n--- Add Student ---")
        student_id = input("Enter ID: ")
        name = input("Enter Name: ")
        subjects_input = input("Enter Subjects (comma separated): ")
        age = int(input("Enter Age: "))
        grade = input("Enter Grade: ")
        dob = input("Enter Date of Birth (dd-mm-yyyy): ")

        subjects = set(subject.strip() for subject in subjects_input.split(","))
        id_dob = (student_id, dob)

        student = {
            "id_dob": id_dob,
            "name": name,
            "subjects": subjects,
            "age": age,
            "grade": grade
        }

        students.append(student)
        print("✅ Student added successfully!")

    elif choice == "2":
        print("\n--- All Students ---")
        if not students:
            print("No records found.")
        else:
            i = 1
            for student in students:
                print(f"\nStudent {i}")
                print(f"ID: {student['id_dob'][0]}")
                print(f"DOB: {student['id_dob'][1]}")
                print(f"Name: {student['name']}")
                print(f"Subjects: {', '.join(student['subjects'])}")
                print(f"Age: {student['age']}")
                print(f"Grade: {student['grade']}")
                i += 1

    elif choice == "3":
        print("\n--- Search Student ---")
        search_id = input("Enter ID to search: ")
        found = False

        for student in students:
            if student["id_dob"][0] == search_id:
                print("✅ Student Found:")
                print(student)
                found = True
                break

        if not found:
            print("❌ Student not found.")

    elif choice == "4":
        print("\n--- Delete Student ---")
        delete_id = input("Enter ID to delete: ")
        found = False

        for student in students:
            if student["id_dob"][0] == delete_id:
                students.remove(student)
                print("✅ Student deleted successfully!")
                found = True
                break

        if not found:
            print("❌ Student not found.")

    elif choice == "5":
        print("\n--- All Subjects ---")
        all_subjects = set()

        for student in students:
            all_subjects.update(student["subjects"])

        if all_subjects:
            print("Subjects:", ", ".join(all_subjects))
        else:
            print("No subjects found.")

    elif choice == "6":
        print("Exiting program...")
        break

    else:
        print("❌ Invalid choice. Try again.")
```
