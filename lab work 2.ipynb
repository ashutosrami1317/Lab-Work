```python
num1 = float(input("Enter frist number:"))
num2 = float(input("Enter second number:"))
op = input("Enter operator (+,-,*,/):")

match op:
    case '+':print("Result:",num1+num2)
    case '-':print("Result:",num1-num2)
    case '*':print("Result:",num1*num2)
    case '/':print("Result:",num1/num2)
```
