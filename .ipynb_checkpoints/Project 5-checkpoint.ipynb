```python
class Employe:
    def __init__(self,name,age,employe_id=None,salary=None):

        self.name = name
        self.age = age
        self.__employe_id = employe_id
        self.__salary = salary

    def get_employe_id(self):
        return self.__employe_id

    def set_employe_id(self,emp_id):
        self.__employe_id = emp_id

    def get_salary(self):
        return self.__salary

    def set_salary(self,salary):
        if salary > 0:
            self.salary = salary
        else:
            print("Salary must be positiver")

    def display(self):
        print(f"ID: {self.__employe_id},\nName:{self.name},\nAge:{self.age},\nSalary:{self.__salary}")




class Manager:
    def __init__(self,name,age,employe_id,salary,department):
```
