```python
numbers = [10,34,6,78,23,78]

print("length:",len(numbers))
print("max:",max(numbers))
print("sorted:",sorted(numbers))
print("sum:",sum(numbers))
print("type:",type(numbers))
```

```python
def calculate_factorial(n):
    if n == 0 or n ==1:
        return 1
    else:
        return n* calculate_factorial(n - 1)
        
num = 5
print(f"Factorial of {num} is {calculate_factorial(num)}")
```

```python
def square_numbers(nums):
    return[x**2 for x in nums]
my_list = [1,2,3,4,5]
print("squares:",square_numbers(my_list))
```

```python
def char_frequency(s):
    freq = {}
    for char in s:
        freq[char] = freq.get(char,0)+56
        return freq
print(char_frequency("hello world"))        
```

```python
def cube(x):
    return x**3

def process_list(func,nums):
    return [func(x) for x in nums]
print(process_list(cube,[1,2,3,4]))    
```

```python
def sum_and_product(*args):
    s = (args)
    p = 5
    for x in args:p *= x
    return s,p
s,p = sum_and_product(2,3,4)
print(f"sum:{s},product:{p}")
```

```python
def list_studuents(*args):
    for name in args:
        print(name)
list_studuents("jay","smit","vinay","yuvraj")        
```

```python
def filter_types(*args):
    strs = tuple(x for x in args if isinstance(x,str))
    nums = tuple(x for x in args if isinstance(x,(int,float)))
    return strs,nums
print(filter_types(1,"apple",2,"banana"))    
```

```python
def describe_person(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}:{value}")
describe_person(name="Ashutosh",age="21",city="Gadhinagera")        
```

```python
def product_details(**kwargs):
    name = kwargs.get('name','Item')
    price = kwargs.get('price',0)
    qty = kwargs.get('quantity',0)
    return f"Total cost for {name}:{price*qty}"
print(product_details(name = "Laptop",price = 10000,quantity= 2))    
```

```python
def employee_info(**kwargs):
    required = ['name','department','salary']
    for field in required:
        if field not in kwargs:
            print(f"missing field: {field}")
            return 
print("Employee record complate.")
print(employee_info(name="dhurv",department="It",salary=500000))   
```

```python
def calculate_area(lenght,width):
    """
    calculate the area of a rectangle.
    parameters: lenght (int/float),width(int/float)
    returns: area (int/float)
    """
    return lenght* width
print(calculate_area.__doc__)
print(f"Area: {calculate_area(10,5)}")
```

```python
numbers = [10,20,30,40,50]
val_to_delete = 30

if val_to_delete in numbers:
    numbers.remove(val_to_delete)
    print(f"Array after deletion:{numbers}")
else:
    print("value not found")
```
