```python
class student:
    def __init__(self,name,marks):
        self.__name = name
        self.__marks = marks

    def calculate_average(self):
        avg = sum(self.__marks) / len(self.__marks)
        return avg
    def display_grade(self):
        avg = self.calculate_average()
        if avg >= 90: grade = "A"
        elif avg >= 75: grade = "B"
        else: grade = "C"
        print(f"Student:{self.__name},\nAverage:{avg:2f},\nGrade:{grade}")

s = student("Devraj",[85,78,56])
s.display_grade()
```

```python
class Base:
    def show(self):
        print("Base class")

class Level1(Base):
    def show(self):
        print("Level 1 class")
        super().show()

class Level2:
    def show(self):
        print("Level 2 class")

class Hybrid(Level1, Level2):
    def show(self):
        print("Hybrid class")
        super().show()

h = Hybrid()
h.show()
```

```python
class User:
    def __init__(self, name, role):
        self.name = name
        self.role = role
        print(f"User {self.name} initialized as {self.role}")

class Admin(User):
    def __init__(self,name,permissions):
        super().__init__(name,role="Administrator")
        self.permissions = permissions
        print(f"Admin permissions: {self.permissions}")

admin_user = Admin("Alice", ["Read","Write","Delete"])
```
